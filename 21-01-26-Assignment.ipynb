1. The Office Door Badge You are programming a smart door for a tech office. The door should only open if the
user provides the correct "Access Code."
Input: User enters a 4-digit code.
Scenario: If the code is  1234 , print "Access Granted. Door Unlocked." Otherwise, print "Access Denied.
Security Alerted."

```python
access_code = 1234
user_input = input("Enter a 4-digit access code to open the door")

if (int(user_input) == access_code):
    print("Access Granted. Door Unlocked")
else:
    print("Access Denied. Security Alerted")
```

## The Traffic Light Sensor A self-driving car needs to interpret a single traffic light color.
Input: User enters "Red", "Yellow", or "Green".
Scenario: * If "Red" "Full Stop."If "Yellow" "Slow down and prepare to stop."If "Green" "Proceed with caution."Anything else "Error: Signal Malfunction!"

```python
signal_color = input("Enter traffic signal color ")
if signal_color == "Red":
    print("Full Stop")
elif signal_color == "Yellow":
    print("Slow Down and prepare to stop")
elif signal_color == "Green":
    print("Proceed with Caution")
else:
    print("Error: Signal Malfunction")  
```

## Movie Theater Ticket Pricing A theater charges based on the age of the customer.
Input: User enters their age.
Scenario:
Age < 5 "Ticket is Free."Age 5 to 17 "Child Ticket: $10."Age 18 to 60 "Adult Ticket: $18."Age > 60 "Senior Ticket: $12."

```python
customer_age = int(input("Enter customer age "))
if customer_age < 5 :
    print("Ticket is Free")
elif customer_age >= 5 and customer_age < 18 :
    print("Child Ticket:$10")
elif customer_age >=18 and customer_age < 60 :
    print("Adult Ticket: $18")
elif customer_age > 60 :
    print("Senior Ticket:$12")
```

## 7. The 7-Day Steps Tracker A fitness app wants to show your activity for the week.
Data: A list of steps:  [4000, 7500, 10000, 2000, 8000, 12000, 5000] .
Scenario: Loop through the list and for each day, print: "Day [X]: You walked [Steps] steps."

```python
no_of_steps = [4000, 7500, 10000, 2000, 8000, 12000, 5000]
i=1
for steps in no_of_steps :
    print("Day",i,":","You walked", steps, "steps")
    i+=1
```

## Functions Example Practice

```python

def add(a,b):
    return(a+b)
def sub(a,b):
    return(a-b)
def mul(a,b):
    return (a*b)
def div(a,b):
    return (a//b)


user_input = int(input("Eneter 1 for Add, 2 for Subsract, 3 for Multiply, 4 for Division"))
a = int(input("Enter the first number >>> "))
b = int(input("Enter the second number >>> "))


if user_input == 1:
    result = add(a,b)
    operator = "+"
elif user_input == 2:
    result = sub(a,b)
    operator = "-"
elif user_input == 3:
    result = mul(a,b)
    operator = "*"
elif user_input == 4:
    result = div(a,b)
    operator = "/"
else:
    print("Enter the valid input")

print("Result is >>> ", a, operator, b, "=", result)
```
