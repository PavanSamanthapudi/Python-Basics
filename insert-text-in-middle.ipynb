```python
def new_func():
    with open("files/dummy.txt", "r+") as f: #opens the file in the context manager a+ mode and pointer will be at the end
        f.seek(0, 2)
        last_index = f.tell()        
        middle_index = int(last_index/2)
        f.seek(middle_index) #move the cursor to the middle index
        second_part = f.read()        
        f.seek(middle_index)        
        text_to_insert = input("Enter the text to be inserted : ")
        f.write(text_to_insert)
        f.write(second_part)

new_func()
    

```
