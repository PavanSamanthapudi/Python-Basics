```python
import sqlite3
```

```python
conn = sqlite3.connect('student_1.db')
```

```python
cursor = conn.cursor()
```

```python
cursor.execute('PRAGMA foreign_keys = ON;')
```

```python
cursor.execute('''
    CREATE TABLE ultimate_data_science2 (
        student_id INT PRIMARY KEY NOT NULL,
        name TEXT NOT NULL,
        batch_id INT NOT NULL
    )
''')
```

```python
student_vals = [(1, "Agi Ann Maria", 2), (2, "Tanmay", 2), (3, "satyajit roy", 2)]
```

```python
for student_val in student_vals:
    print(student_val)
    cursor.execute('''
    INSERT INTO ultimate_data_science2 (student_id, name, batch_id)
    VALUES (?,?,?)
    ''', student_val)
```

```python
conn.commit()
```

```python
conn.close()
```
