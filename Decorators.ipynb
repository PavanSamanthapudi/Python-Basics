```python
def add():
    a=12
    b=13
    c=a+b
    print(c)
```

```python
add()
```

```python
def calc(func):
    def wrapper():
        print("Decorator Started -  add() method")
        func()
        print("Decorator Ended - add() method")
    return wrapper
```

```python
add = calc(add)
add()
```

```python
def add1(a,b):
    c=a+b
    print(c)
```

```python
def calc1(func):
    def wrap(a,b):
        func(a,b)
    return wrap
```

```python
add_1 = calc1(add1)
```

```python
add_1(1,2)
```

```python
def add3(*args, **kwargs):
    print(sum(args))
    vals = kwargs.items()
    print(vals)
    vals_total = 0
    for key, value in vals:
        vals_total += value        
    total = int(sum(args)) + vals_total
    return total
```

```python
def deco(func):
    def wrap(*args, **kwargs):
        print("wrapper started")
        total = func(*args, **kwargs)
        print("wrapper ended")
        return total
    return wrap

```

```python
add4 = deco(add3)
```

```python
add4(1,2,3,a=4,b=5)
```

```python
def deco1(func):
    def wrap(*args, **kwargs):
        print("wrapper started")
        total = func(*args, **kwargs)
        print("wrapper ended")
        return total
    return wrap
```

```python
@deco1
def add5(*args, **kwargs):
    print(sum(args))
    vals = kwargs.items()
    print(vals)
    vals_total = 0
    for key, value in vals:
        vals_total += value        
    total = int(sum(args)) + vals_total
    return total
```

```python
add5(1,3,5,d=9,e=6)
```
