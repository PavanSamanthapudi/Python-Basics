```python
class Motherboard:
    def __init__(self, ram, cpu, chip_type):
        self.ram = ram
        self.cpu = cpu
        self.chip_type = chip_type

class Computer(Motherboard):
    def __init__(self, ram, cpu, chip_type):
        super().__init__(ram, cpu, chip_type)
        self.os = self.__deriveOS()
        

    def __deriveOS(self): #__ means private method
        if self.chip_type.lower() == "mac-chip":
            return "Mac OS"
        else:
            return "other"
        
    def check_my_os(self): # public method
        print("User has :",self.os, "OS")

```

```python
cmp1 =Computer("12gb", "4.2ghz", "mac-chip")
```

```python
cmp1.check_my_os()
```

## Abstraction

```python
from abc import ABC, abstractmethod
```

```python
class AbstractExample(ABC):
    
    @abstractmethod
    def add(self):
        pass

    @abstractmethod
    def sub(self):
        pass
```

```python
class Calc(AbstractExample):
    def add(self):
        print("add")
```

```python
c = Calc()
```

```python
class Calc(AbstractExample):
    def add(self):
        print("add")
    def sub(self):
        print("sub")
```

```python
c = Calc()
```

```python
c.add()
c.sub()
```
