```python
class Manufacturer:
    def __init__(self, maker, car_name, cost):
        self.maker = maker
        self.car_name = car_name
        self.cost = cost
        self.list_of_cars_searched = []
    
    def __str__(self):
        return f"Car Manufacturer is {self.maker}"
    
    def __repr__(self):
        return f"Car manufacturer is {self.maker} and Car model name is {self.car_name}"

    def __add__(self, other):
        return Manufacturer(self.maker+":"+other.maker,self.car_name+":"+other.car_name,self.cost + other.cost)
    
    def __eq__(self, other):
        return self.car_name == other.car_name
    
    def do_search(self):
        self.list_of_cars_searched.append(self.car_name)

    def __len__(self):
        return(len(self.list_of_cars_searched))
    
```

```python
class GettersAndSettersExample:
    def __init__(self, amount):
        self.__balance = amount
    
    @property
    def balance(self):
        print("This is from a Getter property 'balance'")
        return self.__balance
    
    @balance.setter
    def balance(self, value):
        print("This is from a Setter property")
        self.__balance = value
    
```

```python
a = GettersAndSettersExample(1000)
```

```python
a.balance
```

```python
a.balance = 2000
```

```python
a.balance
```

```python
a._GettersAndSettersExample__balance
```

```python
class Account:
    def __init__(self, amount):
        self.__balance = amount
    
    def get_balance(self):
        return self.__balance
    
    def change_balance(self, value):
        print("Changing the balance value from the normal function")
        self.__balance = value
```

```python
a = Account(6000)
```

```python
a.get_balance()
```

```python
a.change_balance(4000)
```

```python
a.get_balance()
```
